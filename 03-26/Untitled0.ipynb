```python
ssudata = {'단과대학':'전문대학', '학과':'AI학과', '전화번호':'5445'}

print(ssudata)
print(ssudata['단과대학'])
print(ssudata['학과'])
print(ssudata['전화번호'])

print(type(ssudata))
```

```python
import json

## ensure_ascii = False 안하면 한글 꺠짐
## json.dumps 데이터 만들떄
jdata = json.dumps(ssudata, ensure_ascii=False)
print(jdata)

# json은 문자열
print(type(jdata))
```

```python
## 데이터를 불러올떄 json.loads
data2 = json.loads(jdata)
print(data2)
print(type(data2))
```

```python
major = [
    {'단과대학':'전문대학', '학과':'AI학과', '전화번호':'5445'},
    {'단과대학':'전문대학', '학과':'방사선학과', '전화번호':'1234'},
    {'단과대학':'전문대학', '학과':'IT학과', '전화번호':'5467'},
]

major
```

```python
print(type(major))
```

```python

jdata2 = json.dumps(major, ensure_ascii=False)
print(jdata2)
print(type(jdata2))
```

```python
data3 = json.loads(jdata2)
print(data3)
print(type(data3))

for i in data3:
  print(i)
  print(type(i))
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/수요일/03-26/전국휴양림표준데이터.json', 'r') as files:
  temp = json.load(files)

temp
```

```python
data = []

for d in temp['records']:
  data.append(d)

data
```

```python
pdData = pd.DataFrame(data)
pdData.head()
```

```python
pdData.to_excel('/content/drive/MyDrive/Colab Notebooks/수요일/03-26/자연휴양림출력화일2.xlsx')
```

```python
import folium
anMap = folium.Map(location=[37.4965, 126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale=True)
anMap
```

```python
for n, row in pdData.iterrows():
  mk = folium.Marker(location=[row['위도'], row['경도']], popup = folium.Popup(row['휴양림명'], max_width=200))
  mk.add_to(anMap)

anMap
```

```python
anMap2 = folium.Map(location=[37.4965, 126.9573],
                   zoom_start=10,
                   zoom_control=False,
                   control_scale=True)
anMap2
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap2)

for n, row in pdData.iterrows():
  folium.Marker(location=[row['위도'], row['경도']], popup = folium.Popup(row['휴양림명'], max_width=200)).add_to(mc)

anMap2
```
